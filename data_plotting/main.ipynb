### **4) DATA PLOTTING**

```python
# Overtime Plot 

# Selected countries filter
selected_countries = ["EU27_2020", "FR", "DE", "ES", "SE", "NL", "HU", "EL"]
country_labels = {
    "EU27_2020": "Euro 27 (2020)",
    "FR": "France",
    "DE": "Germany",
    "ES": "Spain",
    "SE": "Sweden", 
    "NL": "Netherlands",
    "HU": "Hungary",
    "EL": "Greece"
}
gdp_df = country_specific_test_df[
    country_specific_test_df["Country"].isin(selected_countries)
][["Country", "Time", "GDP (Million USD)"]].copy()

# Datetime and pivot selection
gdp_df["Time"] = pd.to_datetime(gdp_df["Time"], errors="coerce")
gdp_pivot = gdp_df.pivot(index="Time", columns="Country", values="GDP (Million USD)")

# We plot the line for each selected country 
plt.figure(figsize=(12, 6))
for country in selected_countries:
    if country in gdp_pivot.columns:
        plt.plot(gdp_pivot.index, gdp_pivot[country], label=country_labels[country], linewidth=2)

# General settings
plt.title("Figure 1 - GDP Over Time per selected EU member states (Million USD)", fontsize=14, weight="bold")
plt.xlabel("Time", fontsize=12)
plt.ylabel("GDP (Million USD)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(title="Country", fontsize=10)
plt.tight_layout()
plt.show()
```
